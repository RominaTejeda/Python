Esta operación se conoce como **conjunción.**
El resultado de una conjunción es verdadero
cuando las dos condiciones que se propone
conectar son verdaderas.
En este caso, y > 1 and y < 10 es verdadero
únicamente cuando y > 1 e y < 10 son ambas

resultado también lo es.


Por otro lado, la **disyunción** nos permite saber si
alguna de dos comparaciones es verdadera.
Una disyunción es verdadera cuando al menos
una de las dos comparaciones que conecta es
verdadera.


````python
y > 1 or y > 15  # solo se cumple y > 1
````

La negación, que se ejecuta con la palabra reservada **not**, invierte el valor lógico de un booleano.


````python
not y < 1 # es 10 no menor a 1?
````

````python
w = num > 5 and num < 7
g = num > 5 or num < 7
k = not num > 5
print("num > 5 y num < 7?", w)
print("num > 5 o num < 7?", g)
print("Es num NO mayor a 5?", k)
````


Al ser las LISTAS colecciones **ordenadas** de objetos, cada elemento tiene un índice, que es la posición en la que se encuentra en la lista empezando desde 0.
A su vez las listas son **mutables,** es decir,  es posible cambiar su contenido.

Cómo acceder a los elementos de una lista en Python?
La **indexación** se puede realizar con uno, dos
o tres parámetros separados por dos puntos



````python
colores = ['azul', 'naranja', 'lila', 'celeste', 'rosado', 'violeta']
````

2.png

````python
colores[1] #Con un parámetro se obtiene un elemento en particular.
#Se empieza a contar desde cero en adelante para seleccionar elementos contando desde la izquierda
````

````python
colores[-2] # y números negativos para seleccionar empezando desde la derecha.
````

````python
colores[7] # que pasa si coloco un indice fuera de rango?
````

El **Slicing** o particionado,consiste en seleccionar **porciones de la lista**.

 Si en lugar de un número escribimos dos  números inicio y fin separados por dos puntos **(inicio:fin)** Python interpretará que queremos una lista que vaya desde la posición inicio a la posición fin, **sin incluir este último**.

 Si escribimos tres números **(inicio:fin:salto)** en lugar de dos, el tercero se utiliza para determinar cada cuantas posiciones añadir un elemento a la lista.

````python
colores[1:4]
# Con dos parámetros se puede seleccionar porciones (slices) de la secuencia.
# El limite inferior se incluye y el limite superior se excluye
````

````python
colores[:2]
# Al no poner primer valor Python asume que es cero o el inicio
````

````python
colores[2:]
# Al no poner segundo valor Python asume que son todos hasta el final
````

````python
colores[0:5:2]
# El tercer parámetro indica el paso con el que los elementos deben tomarse.
````

Entonces obtiene los elementos del 0 hasta el 5, pero ignorando algunos. Comienza en el 0, y avanza 2 pasos, así que selecciona el que tiene el índice 2. Luego, desde el 2 avanza otros 2 pasos y se encuentra con el índice 4

````python
colores[::-1]
#para invertir una lista se tiene que avanzar en pasos de -1,seleccionado desde el inicio hasta el fin
````

# Diccionarios
Son colecciones en donde cada elemento consta de dos partes: **una clave y un valor**.

Los valores pueden ser cualquier tipo de dato, mientras **las claves tienen que ser obligadamente un tipo de dato inmutable.**
Podemos utilizar cualquier objeto inmutable como clave de un diccionario, esto incluye números, strings, tuplas.Además, **las claves no se pueden repetir.**

Los valores correspondientes a una clave se pueden modificar y se puede agregar o quitar elementos.
Los elementos NO están ordenados, con lo cual no se puede acceder por posición ni slicing.

Los diccionarios pueden contener distintos tipos de datos y tambien otras colecciones como listas, tuplas y a hasta otros diccionarios.

Para crear un diccionario vamos a emplear llaves
y separar sus miembros por comas, que a su vez
serán pares de una clave y un valor separados
por dos puntos.

````python
clientes =  {"Apellido":'Herrera', "DNI": 40565999}
````

````python
# Otra forma de definir diccionarios con la funcion dict()
dic2 = dict(
    Nombre="Santiago",
    Apellido="Hernandez",
    Pais="España",
    Ciudad="Madrid"
)
````

````python
print(clientes["Apellido"]) #podemos acceder a un elemento por su clave
````

# Bucle while


El bucle while en particular **repite una porción de
código** siempre que una expresión sea verdadera.

Si la  expresión siempre es verdadera el bucle se repetira infinitamente, y para que un bucle sea útil debe terminar en
algún momento. Hay dos formas de conseguirlo:

1) Que la condición se vuelva falsa (deja de ejecutarse)

2) Ejecutar una instrucción para forzar al bucle a que termine. Cuando Python se encuentra con la instrucción **break** dentro de un bucle, éste termina abruptamente. Break aborta el bucle, independientemente  que la condición sea verdadera.

La sintaxis utilizada para definir la sentencia while es la siguiente:

while <expresión>:
    <sentencia(s)>


<sentencia(s)> es el bloque de sentencias en Python que se ejecutará de manera indefinida hasta que no se cumpla la expresión.

<expresión> es una expresión evaluada en un contexto booleano. El resultado de evaluar esta expresión debe ser True o False. Controla la ejecución del bucle while

El flujo de ejecución de la estructura **`while`** es el siguiente:

1. Determinar si la condición de corte es verdadera o falsa (`True` o `False`).
2. Si la condición es verdadera, se ejecuta el cuerpo dentro del `while` y se vuelve al paso 1.
3. Si la condición es falsa, se sale del bucle sin correr el bloque de código *indentado*, y se continúa ejecutando el programa en la línea siguiente (no *indentada*).

Es importante tener en cuenta que la condición se evalúa al principio del ciclo y no se volverá a evaluar hasta tanto no se terminen de ejecutar todas las instrucciones contenidas dentro del bloque **`while`**.

````python
contador = 1 #variable llamada contador
print('Contando del 1 al 5')
while contador <= 5: #expresión
    print(f'{contador}') #sentencia
    contador += 1    # esto equivale: contador = contador + 1
print('Fin')
````

☝ Para utilizar una f-string, se coloca la letra "f" justo antes de las comillas que delimitan el string y se puede incluir expresiones entre llaves "{}" dentro de la cadena para representar los valores que se desean incluir.

**El paso a paso!**

En el script anterior se inicializa una variable llamada "contador" con el numero 1.
Seguidamente se muestra un mensaje y en la línea 3 aparece una sentencia while. En esta sentencia se comprueba si el valor de la variable contador es menor o igual a 5.
Como se evalúa a True, se muestra por pantalla el numero y después se incrementa el valor de contador en 1.
A continuación, se debería ejecutar el código de la línea 6. Sin embargo, como hemos definido un bucle while, el flujo de ejecución del programa no continúa por la línea 6, sino que vuelve a la línea 3 y de nuevo se evalúa si la expresión contador <= 5 sigue siendo True. En esta ocasión el valor de la variable contador es 2 por lo que la expresión da como resultado True y vuelven a ejecutarse las líneas de la sentencia while. Esto ultimo se repetira hasta que 'contador' sea igual a 6. En esa ocasión la expresión de comparación se evaluará a False y el flujo continuará, ahora sí, por la línea 6 e imprimira por pantalla el mensaje de 'Fin'.

````python
iniciar = 0

while True:
    print("El numero es:", iniciar)
    iniciar += 1

    if iniciar == 5:  # que pasa si no colocaba el break?
        break

````

# Bucle for
Es un bucle con un **numero determinado de iteraciones**, ya que permite iterar sobre cada uno de los elementos de un iterable, como una lista, por ejemplo.
En cada paso de la iteración se tiene en cuenta a un único elemento del objeto iterable, sobre el cuál se pueden aplicar una serie de operaciones.

💡 Un iterable es un objeto que se puede iterar sobre él, es decir, que permite recorrer sus elementos uno a uno. Los tipos de datos string, lista, tupla, diccionario son objetos iterables.

La sintaxis utilizada para definir la sentencia `for` es la siguiente:
```
for <variable> in <iterable>:
    <sentencia(s)>
```

`<iterable>` es una colección de elementos, por ejemplo, una lista o una tupla.

`<variable>` es una variable que recibirá uno de los elementos del objeto `<iterable>` en cada una de las iteraciones.

`<sentencia(s)>` es el bloque de sentencias en Python que se ejecutará repetidamente.

Vamos a utilizar una lista para nuestro ejemplo:

````python
num = [4, 78, 9, 84]  # sobre una lista
for n in num:         # n es la variable que toma el valor del elemento dentro del iterador en cada paso del bucle
    print(n)          # Respetar la indentación para que forme parte del loop
````

````python
#Ejemplo: defino funcion para encontrar los num pares en una lista
def par(numeros):
    lista_num = []
    for i in numeros:
        if i % 2 == 0:
            lista_num.append(i)
    return lista_num
````

✅ El return debe colocarse fuera del ciclo for, de modo que la función devuelva la lista de números pares después de que haya terminado de iterar sobre todos los elementos en la lista de números.
Si colocamos el return dentro del ciclo for, la función devolvería inmediatamente el primer número par que encontrara, sin verificar el resto de la lista.

````python
numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9]
````

````python
# llamo a la función
par(numeros)
````

Ejercicio 9:

Escribir una función que recibe un número y devuelve *True* si el número es primo y *False* en caso contrario.

Mediante un *for* verificar la *primalidad* de los numeros del $1$ al $20$, es decir, decidir si cada número es primo o no.

**Tips:**

* Un número $N$ es primo cuando tiene exactamente $2$ divisores ($1$ y $N$).
* Sin embargo, alcanza con verificar que no es múltiplo de ningún número entre $2$ y $\sqrt{N}$ (recordar que $\sqrt{N}=N^{0.5}$)
* El numero 1 **NO** es primo.

````python
def esPrimo(n):
  if n == 1:
    return False
  d = 2
  while d <= n**0.5:
    if n % d == 0:
      return False
    d = d + 1
  else:
    return True
````

````python
for i in range(1, 21):
  print( i, "Es primo?", esPrimo(i) )
````

# Enviar correo electrónico con el informe de forma automática

````python
from email.message import EmailMessage
import smtplib


# Configuración del servidor SMTP
smtp_host = 'smtp.gmail.com'
smtp_port = 587
emisor = 'tumail@gmail.com'
password = 'contraseña'


# Otras definiciones
destinatarios = ['contacto1', 'contacto2']
asunto = 'Notas Finales del trimestre 2do E'
cuerpo = """
Buenos dias Director
Envio las notas finales de los alumnos de 2do E correspondiente al primer trimestre
Saludos
Paula
"""
ruta_adjunto = ruta_archivo

# Crear el objeto EmailMessage
email = EmailMessage()
# Configurar los encabezados del correo electrónico
email["From"] = emisor
email["To"] = destinatarios
email["Subject"] = asunto

# Contenido del correo electrónico
email.set_content(cuerpo)

# Adjuntar el archivo al correo electrónico
with open(ruta_adjunto, 'rb') as f:
    archivo_adjunto = f.read()
    email.add_attachment(archivo_adjunto, maintype='application',
                       subtype= 'vnd.openxmlformats-officedocument.spreadsheetml.sheet', filename='Notas_1er_trimestre_2do_E.xlsx')

# Para archivos Excel en formato .xlsx (Excel 2007 y versiones posteriores),
# el subtype es 'vnd.openxmlformats-officedocument.spreadsheetml.sheet'.
````

Al utilizar **'rb**', se asegura que el archivo se lea en su forma binaria original, preservando cualquier estructura o contenido específico del archivo. Esto es importante cuando se desea leer o manipular archivos que no están compuestos únicamente por texto plano.

````python
# Conectar y enviar el correo electrónico
with smtplib.SMTP(smtp_host, smtp_port) as sm:
    sm.starttls()
    sm.login(emisor, password)
    sm.send_message(email)
    print('Correo electrónico enviado correctamente.')
````

Este código utiliza la librería smtplib en Python para enviar un correo electrónico utilizando un servidor SMTP.

 SMTP (Simple Mail Transfer Protocol) es un protocolo de red utilizado para el envío de correos electrónicos a través de Internet. Un servidor SMTP es un software o servicio que implementa este protocolo y se encarga de recibir, procesar y entregar los correos electrónicos salientes a sus destinatarios.

**with smtplib.SMTP(smtp_host, smtp_port):** Establece una conexión con el servidor SMTP utilizando el nombre del host y el puerto proporcionados.

**sm.starttls():** Inicia el cifrado TLS (Transport Layer Security) para asegurar la comunicación con el servidor SMTP.

**sm.login(emisor, password):** Inicia sesión en el servidor SMTP utilizando las credenciales del emisor del correo electrónico (dirección de correo y contraseña).

**sm.send_message(email):** Envía el mensaje de correo electrónico especificado, donde email es un objeto de tipo EmailMessage que contiene toda la información del correo electrónico (destinatario, asunto, cuerpo, adjuntos, etc.).
